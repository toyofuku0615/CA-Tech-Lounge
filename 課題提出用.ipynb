## データセットの作成（ベースライン）

データ分析を今後進めていくうえで、精度が上がったか下がったかを判断するために基本となるベースラインをまず作成する。今回は直観的に木々の種類に大きく関係しやすそうな地形情報と土壌情報を組み込んだ。ここから変数を足し引きしたりして精度の向上を目指す

```python
#データフレームに「ID]列を作成
df['ID'] = df.index

# 基本となる地形関連の特徴量を選ぶ。ここれをベースラインとして改良していく
features = [
    "標高",
    "アスペクト",
    "傾斜",
    "Horizontal_Distance_To_Hydrology",
    "Vertical_Distance_To_Hydrology",
    "Horizontal_Distance_To_Roadways",
    "Hillshade_9am",
    "Hillshade_Noon",
    "Hillshade_3pm"
]

x_train = df[features]
y_train = df[["Cover_Type"]]
id_train = df[["ID"]]

print(x_train.shape, y_train.shape, id_train.shape)
```

## バリデーション設定

機械学習モデルは、訓練データを用いてパターンを学習するが、訓練データだけで評価すると過学習が起こり、モデルが訓練データに過剰に適応し、未知のデータに対して正しく予測できなくなる可能性がある。そこで、 機械学習モデルの性能を評価するために、データを学習用（訓練データ）とは別に分割し、モデルの汎化性能を確認するバリデーションを行う必要がある


データセットを訓練データと検証データに分割し、訓練データでモデルを学習した後、検証データを用いて性能を評価する**ホールドアウト検証**と、データセットを k 等分 して、各回ごとに異なるグループを検証データとして使用し、残りを訓練データとして学習を行う。これを k 回繰り返し、すべてのデータが1回ずつ検証データとして使用されるようにする**クロスバリデーション検証**がある。

### ホールドアウト検証

データを8:2に分割する。この時、全体の目的変数のCover_Typeの1～7のクラスの割合と、検証データのCover_Typeの1～7のクラスの割合が等しくなるように分割するよう調整する。stratify=y_trainの部分で調整できている。

```python
from sklearn.model_selection import train_test_split

# 訓練データと検証データに分割
x_tr, x_va, y_tr, y_va, id_tr, id_va = train_test_split(
    x_train, y_train, id_train, test_size=0.2, shuffle=True, stratify=y_train, random_state=123
)

# 形状を確認
print(x_tr.shape, y_tr.shape, id_tr.shape)
print(x_va.shape, y_va.shape, id_va.shape)

# クラス分布の確認
print("y_train クラス分布:\n", y_train.squeeze().value_counts(normalize=True) * 100)
print("y_tr クラス分布:\n", y_tr.squeeze().value_counts(normalize=True) * 100)
print("y_va クラス分布:\n", y_va.squeeze().value_counts(normalize=True) * 100)
```

### クロスバリデーション

n_splits = 5の部分でどのようにテストデータを分類するかを指定する

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold

# クロスバリデーションの設定（5分割）
n_splits = 5
cv = list(StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=123).split(x_train, y_train))

# 各フォールドでのデータ分割を確認
for nfold in np.arange(n_splits):
    print("-"*20, f"Fold {nfold+1}", "-"*20)

    # 訓練用・検証用のインデックス取得
    idx_tr, idx_va = cv[nfold][0], cv[nfold][1]

    # 訓練データ・検証データを分割
    x_tr, y_tr = x_train.iloc[idx_tr, :], y_train.iloc[idx_tr, :]
    x_va, y_va = x_train.iloc[idx_va, :], y_train.iloc[idx_va, :]

    # 分割後のデータ形状を出力
    print(x_tr.shape, y_tr.shape)
    print(x_va.shape, y_va.shape)

    # Cover_Type のクラス分布を表示
    print("y_train クラス分布:\n", y_train["Cover_Type"].value_counts(normalize=True) * 100)
    print("y_tr クラス分布:\n", y_tr["Cover_Type"].value_counts(normalize=True) * 100)
    print("y_va クラス分布:\n", y_va["Cover_Type"].value_counts(normalize=True) * 100)
```

モデル学習

・'objective'を 'multiclass',  に指定することで他クラス分類をとけるようにする

・AUC は二値分類向けなので、多クラス分類では multi_logloss（マルチクラスの交差エントロピー）を使う

・他クラス問題では'metric'を 'multi_logloss'と変更し、'num_class': 7とクラス数を明示的に指定する必要がある

```python
import lightgbm as lgb

# ハイパーパラメータ
params = {
    'boosting_type': 'gbdt',
    'objective': 'multiclass',  # 多クラス分類
    'num_class': 7,  # クラス数を指定
    'metric': 'multi_logloss',  # 多クラス分類の評価指標
    'learning_rate': 0.1,
    'num_leaves': 16,
    'n_estimators': 100000,
    "random_state": 123,
    "importance_type": "gain",
}

# モデル定義
model = lgb.LGBMClassifier(**params)

# モデル学習
model.fit(
    x_tr,
    y_tr.values.ravel(),  # LightGBM の y は 1D にする
    eval_set=[(x_tr, y_tr.values.ravel()), (x_va, y_va.values.ravel())],
    early_stopping_rounds=100,
    verbose=10,
)
```

日本語の列名に対応していないモデルもある
→チャットGPTに列名を英語にして

```python
# 1. 日本語のカラム名を英語に変換する辞書を定義
rename_dict = {
    "標高": "Elevation",
    "アスペクト": "Aspect",
    "傾斜": "Slope",
    "Horizontal_Distance_To_Hydrology": "H_Dist_Hydrology",
    "Vertical_Distance_To_Hydrology": "V_Dist_Hydrology",
    "Horizontal_Distance_To_Roadways": "H_Dist_Roadways",
    "Hillshade_9am": "Hillshade_9am",
    "Hillshade_Noon": "Hillshade_Noon",
    "Hillshade_3pm": "Hillshade_3pm",
    "Horizontal_Distance_To_Fire_Points": "H_Dist_FirePoints",
}

# 2. x_train のカラム名を変更
x_train = x_train.rename(columns=rename_dict)

# 3. x_tr, x_va にも適用
x_tr = x_tr.rename(columns=rename_dict)
x_va = x_va.rename(columns=rename_dict)

# 4. 変更後のカラム名を確認
print("修正後のカラム名:", x_train.columns)
```

'Unnamed: 0'と'ID'は学習に不要な列

```python
x_train = x_train.drop(columns=["Unnamed: 0", "ID"], errors="ignore")
x_tr = x_tr.drop(columns=["Unnamed: 0", "ID"], errors="ignore")
x_va = x_va.drop(columns=["Unnamed: 0", "ID"], errors="ignore")

x_train.columns
```

```python
import lightgbm as lgb

# ハイパーパラメータ
params = {
    'boosting_type': 'gbdt',
    'objective': 'multiclass',  # 多クラス分類
    'num_class': 7,  # クラス数
    'metric': 'multi_logloss',  # 多クラス分類の評価指標
    'learning_rate': 0.1,
    'num_leaves': 16,
    'n_estimators': 100000,
    "random_state": 123,
    "importance_type": "gain",
}

# モデル定義
model = lgb.LGBMClassifier(**params)

# モデル学習（コールバック API を使用）→LightGBM のバージョンによる API 変更によるエラーを修正するために使用した
model.fit(
    x_tr,
    y_tr.values.ravel(),  # LightGBM の y は 1D にする
    eval_set=[(x_tr, y_tr.values.ravel()), (x_va, y_va.values.ravel())],
    callbacks=[
        lgb.early_stopping(stopping_rounds=100, verbose=True),
        lgb.log_evaluation(10),
    ],
)
```

```python
from sklearn.metrics import accuracy_score

# 予測（最も確率の高いクラスを選択）
y_tr_pred = model.predict(x_tr)  # 各サンプルの予測クラス
y_va_pred = model.predict(x_va)  # 各サンプルの予測クラス

# 精度（Accuracy）を計算
metric_tr = accuracy_score(y_tr, y_tr_pred)
metric_va = accuracy_score(y_va, y_va_pred)

# 結果を出力
print("[accuracy] tr: {:.2f}, va: {:.2f}".format(metric_tr, metric_va))
```

データの可視化の段階で、クラス不均衡の問題があるため、正解率（Accuracy）だけでなく、他の評価指標も使うべきだという考察をした
→チャットGPTに「クラス不均衡な他クラス分類に対応した正確な評価を行うことのできる指標を教えて」とお願いした

image.png

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, log_loss

# 予測
y_tr_pred = model.predict(x_tr)  # 各サンプルの予測クラス
y_va_pred = model.predict(x_va)  # 各サンプルの予測クラス

# Accuracy（正解率）
metric_tr_acc = accuracy_score(y_tr, y_tr_pred)
metric_va_acc = accuracy_score(y_va, y_va_pred)

# Precision（適合率） - マクロ平均（全クラスの平均）
metric_tr_prec = precision_score(y_tr, y_tr_pred, average="macro")
metric_va_prec = precision_score(y_va, y_va_pred, average="macro")

# Recall（再現率） - マクロ平均
metric_tr_rec = recall_score(y_tr, y_tr_pred, average="macro")
metric_va_rec = recall_score(y_va, y_va_pred, average="macro")

# F1-score（F1値） - マクロ平均
metric_tr_f1 = f1_score(y_tr, y_tr_pred, average="macro")
metric_va_f1 = f1_score(y_va, y_va_pred, average="macro")

# Log Loss（交差エントロピー損失） - クラス確率を使う
y_tr_pred_proba = model.predict_proba(x_tr)
y_va_pred_proba = model.predict_proba(x_va)
metric_tr_logloss = log_loss(y_tr, y_tr_pred_proba)
metric_va_logloss = log_loss(y_va, y_va_pred_proba)

# 結果を出力
print("[Evaluation Metrics]")
print("[Accuracy]    tr: {:.2f}, va: {:.2f}".format(metric_tr_acc, metric_va_acc))
print("[Precision]   tr: {:.2f}, va: {:.2f}".format(metric_tr_prec, metric_va_prec))
print("[Recall]      tr: {:.2f}, va: {:.2f}".format(metric_tr_rec, metric_va_rec))
print("[F1-score]    tr: {:.2f}, va: {:.2f}".format(metric_tr_f1, metric_va_f1))
print("[Log Loss]    tr: {:.4f}, va: {:.4f}".format(metric_tr_logloss, metric_va_logloss))

# 混同行列を出力
print("\n[Confusion Matrix - Validation]")
print(confusion_matrix(y_va, y_va_pred))
```

**各指標からわかること**

・Accuracy が 0.75→ 全体的な分類精度は高いが、クラスごとのバランスが重要

・Recall（0.60）が低い → クラスの取りこぼしが多い

・Log Loss（0.6013）が高め → 確率の信頼度が低い

**混同行列からわかること**

・クラス 1 が クラス 2 に 10789  件 誤分類されている・クラス 2 の誤分類（9780 件がクラス 1 に分類）

→クラス 1 と 2 は最も頻度が多く、特徴の違いが不明瞭


・クラス 5, 6 で少数派の誤分類がある

→少数派のクラス間の特徴の違いを強調できるような特徴量が必要



① クラス1とクラス2の識別を向上させる
→ これらのクラスの特徴をより明確に区別できるように、新しい特徴量を追加するか、既存の特徴量のエンジニアリングを行う。

② リコール（Recall）の向上
→ 誤分類を減らすために、誤分類の多いクラスのデータを増やす（データのバランス調整）や、損失関数に重みを付ける（クラスごとの重み付け）。

③ 少数派クラスの誤分類を減らす
→ クラス5, 6の誤分類を減らすために、少数派クラスの特徴を強調する新たな特徴量を追加するか、SMOTEなどを使ってデータのバランスを取る。

④ Log Loss の改善（確率の信頼度向上）
→ モデルのキャリブレーション（Platt ScalingやIsotonic Regression）を行い、予測確率を適切に補正する。

 特徴量の追加・修正 + データのバランス調整 + モデルのチューニングを組み合わせて改善を図る！

### **クロスバリデーションの場合**

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold

# LightGBM のパラメータ（多クラス分類対応）
params = {
    'boosting_type': 'gbdt',
    'objective': 'multiclass',  # 多クラス分類
    'num_class': 7,  # クラス数を指定
    'metric': 'multi_logloss',  # 適切な評価指標
    'learning_rate': 0.1,
    'num_leaves': 16,
    'n_estimators': 100000,
    "random_state": 123,
    "importance_type": "gain",
}

metrics = []
imp = pd.DataFrame()

n_splits = 5
cv = list(StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=123).split(x_train, y_train))

for nfold in np.arange(n_splits):
    print("-" * 20, f"Fold {nfold}", "-" * 20)
    idx_tr, idx_va = cv[nfold][0], cv[nfold][1]
    x_tr, y_tr = x_train.loc[idx_tr, :], y_train.loc[idx_tr, :]
    x_va, y_va = x_train.loc[idx_va, :], y_train.loc[idx_va, :]

    print(x_tr.shape, y_tr.shape)
    print(x_va.shape, y_va.shape)

    # クラス分布の確認（修正: Cover_Type を使用）
    print("y_train Cover_Type 分布:\n", y_train["Cover_Type"].value_counts(normalize=True) * 100)
    print("y_tr Cover_Type 分布:\n", y_tr["Cover_Type"].value_counts(normalize=True) * 100)
    print("y_va Cover_Type 分布:\n", y_va["Cover_Type"].value_counts(normalize=True) * 100)

    # モデルの定義
    model = lgb.LGBMClassifier(**params)

    # モデルの学習
    model.fit(
        x_tr,
        y_tr.values.ravel(),
        eval_set=[(x_tr, y_tr.values.ravel()), (x_va, y_va.values.ravel())],
        callbacks=[
            lgb.early_stopping(stopping_rounds=100, verbose=True),
            lgb.log_evaluation(100),
        ],
    )

    # 予測
    y_tr_pred = model.predict(x_tr)
    y_va_pred = model.predict(x_va)

    # 精度（Accuracy）
    metric_tr = accuracy_score(y_tr, y_tr_pred)
    metric_va = accuracy_score(y_va, y_va_pred)
    print("[accuracy] tr: {:.2f}, va: {:.2f}".format(metric_tr, metric_va))
    metrics.append([nfold, metric_tr, metric_va])

    # 特徴量の重要度を取得
    _imp = pd.DataFrame({"col": x_train.columns, "imp": model.feature_importances_, "nfold": nfold})
    imp = pd.concat([imp, _imp], axis=0, ignore_index=True)

# クロスバリデーションの結果を出力
print("-" * 20, "Cross-Validation Results", "-" * 20)
metrics = np.array(metrics)
print(metrics)

print("[cv ] tr: {:.2f}±{:.2f}, va: {:.2f}±{:.2f}".format(
    metrics[:,1].mean(), metrics[:,1].std(),
    metrics[:,2].mean(), metrics[:,2].std(),
))

# 特徴量重要度の平均と標準偏差
imp = imp.groupby("col")["imp"].agg(["mean", "std"])
imp.columns = ["imp", "imp_std"]
imp = imp.reset_index(drop=False)

print("Feature Importance Calculation Done.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, log_loss

# 予測
y_tr_pred = model.predict(x_tr)  # 各サンプルの予測クラス
y_va_pred = model.predict(x_va)  # 各サンプルの予測クラス

# Accuracy（正解率）
metric_tr_acc = accuracy_score(y_tr, y_tr_pred)
metric_va_acc = accuracy_score(y_va, y_va_pred)

# Precision（適合率） - マクロ平均（全クラスの平均）
metric_tr_prec = precision_score(y_tr, y_tr_pred, average="macro")
metric_va_prec = precision_score(y_va, y_va_pred, average="macro")

# Recall（再現率） - マクロ平均
metric_tr_rec = recall_score(y_tr, y_tr_pred, average="macro")
metric_va_rec = recall_score(y_va, y_va_pred, average="macro")

# F1-score（F1値） - マクロ平均
metric_tr_f1 = f1_score(y_tr, y_tr_pred, average="macro")
metric_va_f1 = f1_score(y_va, y_va_pred, average="macro")

# Log Loss（交差エントロピー損失） - クラス確率を使う
y_tr_pred_proba = model.predict_proba(x_tr)
y_va_pred_proba = model.predict_proba(x_va)
metric_tr_logloss = log_loss(y_tr, y_tr_pred_proba)
metric_va_logloss = log_loss(y_va, y_va_pred_proba)

# 結果を出力
print("[Evaluation Metrics]")
print("[Accuracy]    tr: {:.2f}, va: {:.2f}".format(metric_tr_acc, metric_va_acc))
print("[Precision]   tr: {:.2f}, va: {:.2f}".format(metric_tr_prec, metric_va_prec))
print("[Recall]      tr: {:.2f}, va: {:.2f}".format(metric_tr_rec, metric_va_rec))
print("[F1-score]    tr: {:.2f}, va: {:.2f}".format(metric_tr_f1, metric_va_f1))
print("[Log Loss]    tr: {:.4f}, va: {:.4f}".format(metric_tr_logloss, metric_va_logloss))

# 混同行列を出力
print("\n[Confusion Matrix - Validation]")
print(confusion_matrix(y_va, y_va_pred))
```

```python
# データフレームに「ID」列を作成
df['ID'] = df.index

# 基本の地形関連の特徴量
base_features = [
    "Elevation",
    "Aspect",
    "Slope",
    "Horizontal_Distance_To_Hydrology",
    "Vertical_Distance_To_Hydrology",
    "Horizontal_Distance_To_Roadways",
    "Hillshade_9am",
    "Hillshade_Noon",
    "Hillshade_3pm",
    "Horizontal_Distance_To_Fire_Points"
]

# Soil_Type1 から Soil_Type40 までの列名をリストに追加
soil_features = [f"Soil_Type{i}" for i in range(1, 41)]

# 全ての特徴量を統合
features = base_features + soil_features

# 特徴量データフレームの作成
x_train = df[features]
y_train = df[["Cover_Type"]]
id_train = df[["ID"]]

# 形状の確認
print(x_train.shape, y_train.shape, id_train.shape)
```

```python
columns_to_log_transform = [
    "Horizontal_Distance_To_Hydrology",
    "Vertical_Distance_To_Hydrology",
    "Horizontal_Distance_To_Roadways",
    "Horizontal_Distance_To_Fire_Points"
]

n_cols = len(columns_to_log_transform)
fig, axes = plt.subplots(1, n_cols, figsize=(5 * n_cols, 5))

for i, col in enumerate(columns_to_log_transform):
    axes[i].hist(x_train[col], bins=50, edgecolor="black")
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```

```python


# ログ変換を適用
columns_to_log_transform = [
    "Horizontal_Distance_To_Hydrology",
    "Vertical_Distance_To_Hydrology",
    "Horizontal_Distance_To_Roadways",
    "Horizontal_Distance_To_Fire_Points"
]

for col in columns_to_log_transform:
    min_val = x_train[col].min()
    if min_val < 0:
        offset = abs(min_val) + 1e-6  # わずかなオフセットを加える
        x_train[col] = np.log1p(x_train[col] + offset)
    else:
        x_train[col] = np.log1p(x_train[col])

# ヒストグラムを横一列に並べて表示（対数化後）
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

for ax, col in zip(axes, columns_to_log_transform):
    ax.hist(x_train[col], bins=30, alpha=0.7, edgecolor='black')
    ax.set_title(f"Log Transformed {col}")
    ax.set_xlabel("Log Distance")
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

#LightGBMのハイパーパラメーターのチューニング(0.8825)+5%

チャットGPTに「精度を改善するために調整することのできるハイパーパラメーターは何がある？」と質問した

LightGBMには、モデルの性能や学習過程に影響を与える多くのハイパーパラメーターが存在します。以下は主要なパラメーターの一例です。

- **boosting_type**  
  - 使用するブースティングアルゴリズムを指定します。  
  - 例: `gbdt`（勾配ブースティング）、`dart`（Dropouts meet Multiple Additive Regression Trees）、`goss`（Gradient-based One-Side Sampling）など。

- **num_leaves**  
  - 1つの決定木における葉（リーフ）の最大数。  
  - 値を大きくするとモデルは複雑になりますが、過学習のリスクも増加します。

- **max_depth**  
  - 決定木の最大の深さ。  
  - 制限を設けることで過学習を防止できます。`-1`の場合は制限なし。

- **learning_rate**  
  - 各反復ステップでの重み更新の大きさを決めます。  
  - 小さい値にすると学習が安定しますが、収束までに時間がかかる可能性があります。

- **n_estimators**（または**num_iterations**）  
  - 決定木（ブースティングラウンド）の数。  
  - 多く設定するとモデルの表現力が上がりますが、計算コストが増大します。

- **min_child_samples**（または**min_data_in_leaf**）  
  - リーフノードに必要な最小のデータ数。  
  - この値が大きいと、細かい分割を抑え、過学習を防ぎます。

- **min_child_weight**  
  - リーフに含まれるサンプルの重みの合計の最小値。  
  - ノイズの多いデータに対して安定性を向上させます。

- **subsample**（または**bagging_fraction**）  
  - 各イテレーションで使用するデータの割合。  
  - サブサンプリングを行うことで、過学習を抑制できます。

- **bagging_freq**  
  - バギング（サブサンプリング）を適用する頻度（ステップ数の間隔）。  
  - 例えば、`bagging_freq=5`なら5回ごとにサブサンプリングを実施します。

- **colsample_bytree**（または**feature_fraction**）  
  - 各決定木を構築する際に使用する特徴量の割合。  
  - 特徴量の一部だけを用いることで、モデルの汎化性能を向上させます。

- **reg_alpha**（L1正則化）  
  - 重みに対するL1正則化の係数。  
  - 過学習を防ぐために使用します。

- **reg_lambda**（L2正則化）  
  - 重みに対するL2正則化の係数。  
  - 同様に、モデルの過学習抑制に役立ちます。

- **min_split_gain**  
  - ノードを分割するために必要な最小の損失改善量。  
  - この値以上の改善がなければ分割を行いません。

- **max_bin**  
  - 特徴量の値をビン分割する際のビンの最大数。  
  - ビン数を調整することで、学習速度と精度のトレードオフを調整できます。

- **objective**  
  - 学習タスクの目的関数を指定します。  
  - 回帰、分類、ランキングなど、目的に応じた関数を設定します。

- **metric**  
  - モデル評価のための指標を指定します。  
  - 例: `rmse`（回帰の場合）、`binary_logloss` や `auc`（2値分類の場合）、`multi_logloss`（多クラス分類の場合）など。

また、boosting_typeが`dart`の場合は、以下のような追加パラメーターがあります。

- **drop_rate**  
  - ドロップするブースティングラウンドの割合。

- **skip_drop**  
  - 分割ドロップの確率を調整するパラメーター。

このように、LightGBMは多様なハイパーパラメーターを提供しており、データの特性やタスクに応じて最適な設定を選択することが可能です。これらのパラメーターのチューニングにより、モデルの精度や学習速度、汎化性能を大幅に向上させることができます。

チューニングには手動チューニングと自動チューニングがある

**手動チューニング**

・試行回数は少ないが簡単に調整できて効率的

**自動チューニング**

・手動チューニングより精度は高くなる

・学習時間が長くなってしまう

・過学習をしてしまう可能性がある

```python
import lightgbm as lgb
import optuna
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
import numpy as np

# 探索しないハイパーパラメータ
params_base = {
    "boosting_type": "gbdt",
    "objective": "multiclass",
    "num_class": 7,
    "metric": "multi_logloss",
    "learning_rate": 0.02,
    "n_estimators": 500,
    "seed": 123,
    "n_jobs": -1,
    "device": "cpu"
}

def objective(trial):
    params_tuning = {
        "num_leaves": trial.suggest_int("num_leaves", 8, 64),  # 制限を強化
        "min_data_in_leaf": trial.suggest_int("min_data_in_leaf", 5, 50),  # 小さく調整
        "min_sum_hessian_in_leaf": trial.suggest_float("min_sum_hessian_in_leaf", 1e-3, 1e-1, log=True),  # 安定化
        "feature_fraction": trial.suggest_float("feature_fraction", 0.7, 0.9),  # サブサンプリングの範囲調整
        "bagging_fraction": trial.suggest_float("bagging_fraction", 0.7, 0.9),  # 同上
        "lambda_l1": trial.suggest_float("lambda_l1", 0.01, 10.0, log=True),
        "lambda_l2": trial.suggest_float("lambda_l2", 0.01, 10.0, log=True)
    }
    params_tuning.update(params_base)

    list_metrics = []
    cv = list(StratifiedKFold(n_splits=3, shuffle=True, random_state=123).split(x_train, y_train))  # n_splits=3 で時間短縮
    for nfold in range(3):
        idx_tr, idx_va = cv[nfold]
        x_tr, y_tr = x_train.loc[idx_tr, :], y_train.loc[idx_tr, :].values.ravel()
        x_va, y_va = x_train.loc[idx_va, :], y_train.loc[idx_va, :].values.ravel()

        model = lgb.LGBMClassifier(**params_tuning)
        model.fit(
            x_tr, y_tr,
            eval_set=[(x_tr, y_tr), (x_va, y_va)],
            callbacks=[
                lgb.early_stopping(stopping_rounds=20, verbose=False),
                lgb.log_evaluation(0)
            ]
        )

        y_va_pred = model.predict(x_va)
        metric_va = accuracy_score(y_va, y_va_pred)
        list_metrics.append(metric_va)

    return np.mean(list_metrics)

sampler = optuna.samplers.TPESampler(seed=123)
study = optuna.create_study(sampler=sampler, direction="maximize")
study.optimize(objective, n_trials=5)

print("Best trial:")
print(study.best_trial)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization, Embedding, Flatten, Concatenate
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# 目的変数
target_column = "Cover_Type"

# 特徴量の選択
num_features = ["Elevation", "Slope", "Hillshade_mean"]  # 数値特徴量のリスト
cat_features = ["Soil_Type1", "Soil_Type2", "Soil_Type3"]  # カテゴリ特徴量のリスト

# 数値特徴量・カテゴリ特徴量を抽出
x_train_num = df[num_features]
x_train_cat = df[cat_features]
y_train = df[target_column]

# データ分割
x_tr_num, x_va_num, x_tr_cat, x_va_cat, y_tr, y_va = train_test_split(
    x_train_num, x_train_cat, y_train, test_size=0.2, stratify=y_train, random_state=123
)

# One-Hot Encoding
y_tr_ohe = to_categorical(y_tr - 1, num_classes=7)
y_va_ohe = to_categorical(y_va - 1, num_classes=7)

# 数値特徴量の前処理（欠損値補完 + 正規化）
scaler = MinMaxScaler()
x_tr_num = scaler.fit_transform(x_tr_num.fillna(x_tr_num.mean()))
x_va_num = scaler.transform(x_va_num.fillna(x_tr_num.mean()))

# モデル作成
def create_model_embedding(input_shape_num, input_shape_cat, unique_categories):
    ################# 数値特徴量
    input_num = Input(shape=(input_shape_num,))
    layer_num = Dense(10, activation="relu")(input_num)
    layer_num = BatchNormalization()(layer_num)
    layer_num = Dropout(0.2)(layer_num)
    layer_num = Dense(10, activation="relu")(layer_num)

    ################# カテゴリ特徴量
    input_cat = Input(shape=(input_shape_cat,))
    layer_cat = Embedding(input_dim=unique_categories + 1, output_dim=min(50, unique_categories // 2))(input_cat)
    layer_cat = Flatten()(layer_cat)
    layer_cat = Dropout(0.2)(layer_cat)

    ################# 結合層
    hidden_layer = Concatenate()([layer_num, layer_cat])
    hidden_layer = Dense(50, activation="relu")(hidden_layer)
    hidden_layer = BatchNormalization()(hidden_layer)
    hidden_layer = Dropout(0.1)(hidden_layer)
    hidden_layer = Dense(20, activation="relu")(hidden_layer)
    hidden_layer = BatchNormalization()(hidden_layer)
    hidden_layer = Dropout(0.1)(hidden_layer)
    output_layer = Dense(7, activation="softmax")(hidden_layer)  # クラス数が 7 なので softmax

    model = Model(inputs=[input_num, input_cat], outputs=output_layer)

    model.compile(
        optimizer="Adam",
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )

    return model

# モデル作成
unique_categories = x_train_cat.nunique().max()
model = create_model_embedding(input_shape_num=x_tr_num.shape[1], input_shape_cat=x_tr_cat.shape[1], unique_categories=unique_categories)

# モデルのサマリー
model.summary()

```

```python
#データフレームに「ID]列を作成
df['ID'] = df.index

# 基本となる地形関連の特徴量を選ぶ。ここれをベースラインとして改良していく
features = [
    "Elevation",
    "Aspect",
    "Slope",
    "Horizontal_Distance_To_Hydrology",
    "Vertical_Distance_To_Hydrology",
    "Horizontal_Distance_To_Roadways",
    "Hillshade_9am",
    "Hillshade_Noon",
    "Hillshade_3pm"
]

x_train = df[features]
y_train = df[["Cover_Type"]]
id_train = df[["ID"]]

print(x_train.shape, y_train.shape, id_train.shape)
```

```python
import lightgbm as lgb
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# パラメータ
params = {
    'boosting_type': 'gbdt',
    'objective': 'multiclass',
    'metric': 'multi_logloss',
    'learning_rate': 0.1,
    'num_leaves': 16,
    'n_estimators': 100000,
    "random_state": 123,
    "importance_type": "gain",
    "num_class": 7,
}

def train_holdout(input_x, input_y, params, test_size=0.2):
    """
    ホールドアウト法で LightGBM を学習し、評価する関数

    Returns:
        model (LGBMClassifier): 学習済みモデル
        imp (pd.DataFrame): 特徴量の重要度
        x_va (pd.DataFrame): 検証用データ
        y_va (pd.Series): 検証用ラベル
        y_va_pred_proba (np.array): 検証データの予測確率（アンサンブル用）
    """

    # データ分割（ホールドアウト）
    x_tr, x_va, y_tr, y_va = train_test_split(input_x, input_y, test_size=test_size, stratify=input_y, random_state=123)

    print(f"Train Shape: {x_tr.shape}, {y_tr.shape}")
    print(f"Validation Shape: {x_va.shape}, {y_va.shape}")

    # モデルの作成
    model = lgb.LGBMClassifier(**params)

    # モデルの学習
    model.fit(
        x_tr, y_tr,
        eval_set=[(x_tr, y_tr), (x_va, y_va)],
        callbacks=[
            lgb.early_stopping(stopping_rounds=100, verbose=True),
            lgb.log_evaluation(100),
        ]
    )

    # 予測確率（アンサンブル用）
    y_va_pred_proba = model.predict_proba(x_va)

    # 精度計算
    y_tr_pred = model.predict(x_tr)
    y_va_pred = model.predict(x_va)
    metric_tr = accuracy_score(y_tr, y_tr_pred)
    metric_va = accuracy_score(y_va, y_va_pred)

    print(f"[Accuracy] Train: {metric_tr:.2f}, Validation: {metric_va:.2f}")

    # 特徴量の重要度
    imp = pd.DataFrame({"Feature": input_x.columns, "Importance": model.feature_importances_})

    print("Training Completed.")
    return model, imp, x_va, y_va, y_va_pred_proba


```

この手法は、「確率的にどのクラスに属する可能性が高いか」を考慮するため、単純な多数決アンサンブル（最も予測回数の多いクラスを選ぶ）よりも滑らかで安定した予測ができる。

### メタ特徴量を追加（0.8722)

追加されたメタ特徴量
特徴量	説明
1. 予測確率	各モデルの (サンプル数, クラス数) の確率分布
2. 予測ラベル	各モデルの (サンプル数, 1) のラベル
3. 予測ラベルの One-Hot Encoding	(サンプル数, クラス数) の One-Hot ベクトル
4. 予測確率の最大値	(サンプル数, 1) の「最も高い確率」
5. 予測確率のエントロピー	(サンプル数, 1) の「不確実性指標」
6. 予測確率の分散	(サンプル数, 1) の「確率分布のばらつき」
7. 予測確率の中央値	(サンプル数, 1) の「中央値」
8. モデル間の予測一致率	(サンプル数, 1) の「他モデルと一致した割合」

```python
import numpy as np
import pandas as pd
import os
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from scipy.stats import entropy

# 保存先ディレクトリ
save_path = "/content/drive/MyDrive/サイバーエージェント/"

# 予測確率とクラスラベルを保存したファイルのリスト
models_proba_list = [
    "LightGBM_Elevation_Slope_予測確率.npy",
    "LightGBM_Hillshade_mean_予測確率.npy",
    "LightGBM_Hillshade_range_予測確率.npy",
    "LightGBM_Horizontal_Distance_To_Fire_PointsとSoil_Typeを追加_予測確率.npy",
    "LightGBM_Horizontal_Distance_To_Fire_PointsとWilderness_Areaを追加_予測確率.npy",
    "LightGBM_Horizontal_Distance_To_Fire_Pointsを追加_予測確率.npy",
    "LightGBM_Hydro_Road_diffを追加_予測確率.npy",
    "LightGBM_Soil_Typeを追加_予測確率.npy",
    "LightGBM_Wilderness_Areaを追加_予測確率.npy",
    "LightGBM_スケール化_予測確率.npy",
    "LightGBM_ベースライン_予測確率.npy",
    "LightGBM_全部の特徴量_予測確率.npy",
    "LightGBM_対数化_予測確率.npy",
    "LightGBM_方位角のsincos変換_予測確率.npy",
    "ニューラルネットワーク_予測確率.npy",
    "ニューラルネットワーク_埋め込み層_予測確率.npy",
    "ロジスティック回帰_予測確率.npy",
    "全データ_ニューラルネットワーク_予測確率.npy"
]

models_label_list = [p.replace("予測確率.npy", "クラスラベル.npy") for p in models_proba_list]

# 各モデルの予測確率と予測ラベルを読み込む
y_va_probas = [np.load(os.path.join(save_path, model), allow_pickle=True) for model in models_proba_list]
y_va_labels = [np.load(os.path.join(save_path, model), allow_pickle=True) for model in models_label_list]

# 正解ラベルを読み込む
y_va_true = np.load(os.path.join(save_path, "LightGBM_ベースライン_正解ラベル.npy"))

# 追加するメタ特徴量の計算
meta_features = []

for proba, label in zip(y_va_probas, y_va_labels):
    # 1. 予測確率（既存）
    meta_features.append(proba)

    # 2. 予測ラベル
    meta_features.append(label.reshape(-1, 1))  # (サンプル数, 1)

    # 3. 予測ラベルの One-Hot Encoding
    num_classes = proba.shape[1]
    one_hot = np.eye(num_classes)[label - 1]  # クラス 1~7 を 0~6 に変換
    meta_features.append(one_hot)

    # 4. 予測確率の最大値（信頼度）
    max_confidence = np.max(proba, axis=1).reshape(-1, 1)
    meta_features.append(max_confidence)

    # 5. 予測確率のエントロピー（不確実性）
    uncertainty = entropy(proba.T).reshape(-1, 1)
    meta_features.append(uncertainty)

    # 6. 予測確率の分散
    variance = np.var(proba, axis=1).reshape(-1, 1)
    meta_features.append(variance)

    # 7. 予測確率の中央値
    median = np.median(proba, axis=1).reshape(-1, 1)
    meta_features.append(median)

# 8. モデル間の予測一致率
y_va_labels_array = np.array(y_va_labels)
agreement = (np.mean(y_va_labels_array == y_va_labels_array[0], axis=0)).reshape(-1, 1)  # 最初のモデルとの一致率
meta_features.append(agreement)

# すべてのメタ特徴量を結合
meta_features = np.hstack(meta_features)  # (サンプル数, 拡張した特徴量数)

# クラスラベルを 0-indexed に変換（XGBoost の仕様に合わせる）
y_va_true_adjusted = y_va_true - 1

# モデル情報の確認
print(f"メタ特徴量の形状: {meta_features.shape}")
print(f"正解ラベルの形状: {y_va_true.shape}")

# メタモデル（XGBoost）を学習
meta_model = XGBClassifier(
    n_estimators=100,
    max_depth=3,
    learning_rate=0.1,
    use_label_encoder=False,
    eval_metric="mlogloss"
)
meta_model.fit(meta_features, y_va_true_adjusted)  # 修正したラベルを使用

# 予測
y_va_ensemble_stacked = meta_model.predict(meta_features) + 1  # 元のクラス (1~7) に戻す

# 精度の評価
stacked_acc = accuracy_score(y_va_true, y_va_ensemble_stacked)
print(f"\n【XGBoost使用】拡張メタ特徴量を用いたスタッキングアンサンブルの精度: {stacked_acc:.4f}")
```

### ハイパーパラメーターを調整する（0.9385）

進捗を見えるようにする

```python
import optuna
import numpy as np
import xgboost as xgb
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import accuracy_score

# meta_features と y_va_true_adjusted は既に作成済みである前提です。
# もし y_va_true_adjusted が 2D の場合は、1D に変換してください。
y_va_true_adjusted = y_va_true_adjusted.ravel()

# 目的関数：交差検証で accuracy を評価指標としています。
def objective(trial):
    params = {
        'objective': 'multi:softprob',
        'eval_metric': 'mlogloss',
        'num_class': len(np.unique(y_va_true_adjusted)),
        'n_estimators': trial.suggest_int('n_estimators', 50, 200),
        'max_depth': trial.suggest_int('max_depth', 2, 6),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'gamma': trial.suggest_float('gamma', 0, 5)
    }

    model = xgb.XGBClassifier(**params, use_label_encoder=False)
    cv = StratifiedKFold(n_splits=2, shuffle=True, random_state=123)
    scores = cross_val_score(model, meta_features, y_va_true_adjusted, cv=cv, scoring='accuracy')
    return np.mean(scores)

# コールバック関数：各試行終了時に結果を出力します。
def callback(study, trial):
    print(f"Trial {trial.number} finished with value: {trial.value}")

# Optuna スタディの作成と最適化（試行回数は n_trials=5 としています）
sampler = optuna.samplers.TPESampler(seed=123)
study = optuna.create_study(direction='maximize', sampler=sampler)
study.optimize(objective, n_trials=5, callbacks=[callback])

print("Best hyperparameters:", study.best_params)

# 最適なハイパーパラメータで最終モデルを学習・評価
final_model = xgb.XGBClassifier(**study.best_params, use_label_encoder=False, eval_metric='mlogloss')
final_model.fit(meta_features, y_va_true_adjusted)
y_pred = final_model.predict(meta_features) + 1  # XGBoostは0-indexedなので元に戻す
acc = accuracy_score(y_va_true, y_pred)
print(f"Final ensemble accuracy: {acc:.4f}")
```

```python
import numpy as np
import pandas as pd
import os
from sklearn.metrics import accuracy_score
import lightgbm as lgb

# 保存先ディレクトリ
save_path = "/content/drive/MyDrive/サイバーエージェント/"

# 予測確率を保存したファイルのリスト
models_proba_list = [
    "LightGBM_Elevation_Slope_予測確率.npy",
    "LightGBM_Hillshade_mean_予測確率.npy",
    "LightGBM_Hillshade_range_予測確率.npy",
    "LightGBM_Horizontal_Distance_To_Fire_PointsとSoil_Typeを追加_予測確率.npy",
    "LightGBM_Horizontal_Distance_To_Fire_PointsとWilderness_Areaを追加_予測確率.npy",
    "LightGBM_Horizontal_Distance_To_Fire_Pointsを追加_予測確率.npy",
    "LightGBM_Hydro_Road_diffを追加_予測確率.npy",
    "LightGBM_Soil_Typeを追加_予測確率.npy",
    "LightGBM_Wilderness_Areaを追加_予測確率.npy",
    "LightGBM_スケール化_予測確率.npy",
    "LightGBM_ベースライン_予測確率.npy",
    "LightGBM_全部の特徴量_予測確率.npy",
    "LightGBM_対数化_予測確率.npy",
    "LightGBM_方位角のsincos変換_予測確率.npy",
    "ニューラルネットワーク_予測確率.npy",
    "ニューラルネットワーク_埋め込み層_予測確率.npy",
    "ロジスティック回帰_予測確率.npy",
    "全データ_ニューラルネットワーク_予測確率.npy"
]

# 正解ラベルの読み込み（1-indexed と仮定）
y_va_true = np.load(os.path.join(save_path, "LightGBM_ベースライン_正解ラベル.npy"))

# 各モデルの精度を計算して表示
model_accuracies = {}
print("各モデルの精度:")
for model_name in models_proba_list:
    proba = np.load(os.path.join(save_path, model_name), allow_pickle=True)
    # 各モデルの予測ラベルは、各行の最大確率のインデックス（0-indexed）に +1 して取得
    y_va_pred = np.argmax(proba, axis=1) + 1
    acc = accuracy_score(y_va_true, y_va_pred)
    model_accuracies[model_name] = acc
    print(f"{model_name}: Accuracy = {acc:.4f}")

# 閾値を設定し、高精度モデルのみを選択（ここでは精度が 0.79 以上のモデルを使用）
accuracy_threshold = 0.79
selected_models = [model for model, acc in model_accuracies.items() if acc >= accuracy_threshold]
print(f"\n⚡ 選択されたモデル（閾値 {accuracy_threshold} 以上）: {len(selected_models)} / {len(models_proba_list)}\n")

# 選択されたモデルの予測確率を読み込み、横方向に結合してメタ特徴量を作成
selected_probas = [np.load(os.path.join(save_path, model), allow_pickle=True) for model in selected_models]
meta_features = np.hstack(selected_probas)
print(f"メタ特徴量の形状 (選択後): {meta_features.shape}")

# LightGBM のメタモデルは 0-indexed を前提とするため、正解ラベルを変換
y_va_true_adjusted = y_va_true - 1

# メタモデルとして LightGBM を学習（パラメータは例示）
meta_model = lgb.LGBMClassifier(
    n_estimators=100,
    max_depth=3,
    learning_rate=0.1,
    min_data_in_leaf=1,    # より細かい分割を許容
    num_leaves=15,         # max_depth=3 の場合、2^3=8～15 程度に設定
    use_label_encoder=False,
    eval_metric="mlogloss",
    verbose=-1             # ログ出力を抑制
)

meta_model.fit(meta_features, y_va_true_adjusted)
# 予測（出力は 0-indexed なので、元のラベルに戻すため +1）
y_va_ensemble_stacked = meta_model.predict(meta_features) + 1

stacked_acc = accuracy_score(y_va_true, y_va_ensemble_stacked)
print(f"\n🎯 【低精度モデル除外後】LightGBM を用いたスタッキングアンサンブルの精度: {stacked_acc:.4f}")
```

### メタ特徴量を追加（ 0.8645)


```python
import numpy as np
import pandas as pd
import os
from sklearn.metrics import accuracy_score
import lightgbm as lgb
from scipy.stats import entropy

# 保存先ディレクトリ
save_path = "/content/drive/MyDrive/サイバーエージェント/"

# 予測確率とクラスラベルを保存したファイルのリスト
models_proba_list = [
    "LightGBM_Elevation_Slope_予測確率.npy",
    "LightGBM_Hillshade_mean_予測確率.npy",
    "LightGBM_Hillshade_range_予測確率.npy",
    "LightGBM_Horizontal_Distance_To_Fire_PointsとSoil_Typeを追加_予測確率.npy",
    "LightGBM_Horizontal_Distance_To_Fire_PointsとWilderness_Areaを追加_予測確率.npy",
    "LightGBM_Horizontal_Distance_To_Fire_Pointsを追加_予測確率.npy",
    "LightGBM_Hydro_Road_diffを追加_予測確率.npy",
    "LightGBM_Soil_Typeを追加_予測確率.npy",
    "LightGBM_Wilderness_Areaを追加_予測確率.npy",
    "LightGBM_スケール化_予測確率.npy",
    "LightGBM_ベースライン_予測確率.npy",
    "LightGBM_全部の特徴量_予測確率.npy",
    "LightGBM_対数化_予測確率.npy",
    "LightGBM_方位角のsincos変換_予測確率.npy",
    "ニューラルネットワーク_予測確率.npy",
    "ニューラルネットワーク_埋め込み層_予測確率.npy",
    "ロジスティック回帰_予測確率.npy",
    "全データ_ニューラルネットワーク_予測確率.npy"
]

# クラスラベルのファイル名は予測確率のファイル名を置換して生成
models_label_list = [p.replace("予測確率.npy", "クラスラベル.npy") for p in models_proba_list]

# 各モデルの予測確率と予測ラベルを読み込む
y_va_probas = [np.load(os.path.join(save_path, model), allow_pickle=True) for model in models_proba_list]
y_va_labels = [np.load(os.path.join(save_path, model), allow_pickle=True) for model in models_label_list]

# 正解ラベルを読み込む（ここでは1-indexed と仮定）
y_va_true = np.load(os.path.join(save_path, "LightGBM_ベースライン_正解ラベル.npy"))

# 追加するメタ特徴量の計算
meta_features_list = []

for proba, label in zip(y_va_probas, y_va_labels):
    # 1. 予測確率（既存）
    meta_features_list.append(proba)

    # 2. 予測ラベル（列ベクトルに変換）
    meta_features_list.append(label.reshape(-1, 1))

    # 3. 予測ラベルの One-Hot Encoding（クラス1～7 を 0～6 に変換）
    num_classes = proba.shape[1]
    one_hot = np.eye(num_classes)[label - 1]
    meta_features_list.append(one_hot)

    # 4. 予測確率の最大値（信頼度）
    max_confidence = np.max(proba, axis=1).reshape(-1, 1)
    meta_features_list.append(max_confidence)

    # 5. 予測確率のエントロピー（不確実性）
    uncertainty = entropy(proba.T).reshape(-1, 1)
    meta_features_list.append(uncertainty)

    # 6. 予測確率の分散
    variance = np.var(proba, axis=1).reshape(-1, 1)
    meta_features_list.append(variance)

    # 7. 予測確率の中央値
    median = np.median(proba, axis=1).reshape(-1, 1)
    meta_features_list.append(median)

# 8. モデル間の予測一致率：最初のモデルとの一致率を算出
y_va_labels_array = np.array(y_va_labels)
agreement = (np.mean(y_va_labels_array == y_va_labels_array[0], axis=0)).reshape(-1, 1)
meta_features_list.append(agreement)

# すべてのメタ特徴量を結合
meta_features = np.hstack(meta_features_list)  # (サンプル数, 拡張した特徴量数)

# クラスラベルを 0-indexed に変換（LightGBM の仕様に合わせる）
y_va_true_adjusted = y_va_true - 1

# モデル情報の確認
print(f"メタ特徴量の形状: {meta_features.shape}")
print(f"正解ラベルの形状: {y_va_true.shape}")

# メタモデル（LightGBM）を学習
meta_model = lgb.LGBMClassifier(
    n_estimators=100,
    max_depth=3,
    learning_rate=0.1,
    min_data_in_leaf=1,
    num_leaves=15,
    use_label_encoder=False,
    eval_metric="mlogloss",
    verbose=-1
)
meta_model.fit(meta_features, y_va_true_adjusted)

# 予測（出力は 0-indexed なので、元のクラス(1～7) に戻すため +1）
y_va_ensemble_stacked = meta_model.predict(meta_features) + 1

# 精度の評価
stacked_acc = accuracy_score(y_va_true, y_va_ensemble_stacked)
print(f"\n【LightGBM使用】拡張メタ特徴量を用いたスタッキングアンサンブルの精度: {stacked_acc:.4f}")
```

```python
import optuna
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import accuracy_score

# meta_features と y_va_true_adjusted は既に作成済みである前提です。
# もし y_va_true_adjusted が 2D の場合は、1D に変換してください。
y_va_true_adjusted = y_va_true_adjusted.ravel()

# 目的関数：交差検証で accuracy を評価指標としています。
def objective(trial):
    params = {
        'objective': 'multiclass',
        'metric': 'multi_logloss',
        'num_class': len(np.unique(y_va_true_adjusted)),
        'n_estimators': trial.suggest_int('n_estimators', 50, 200),
        'max_depth': trial.suggest_int('max_depth', 2, 6),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'bagging_fraction': trial.suggest_float('subsample', 0.6, 1.0),       # subsample → bagging_fraction
        'feature_fraction': trial.suggest_float('colsample_bytree', 0.6, 1.0),  # colsample_bytree → feature_fraction
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'min_split_gain': trial.suggest_float('gamma', 0, 5)                    # gamma → min_split_gain
    }

    model = lgb.LGBMClassifier(**params)
    cv = StratifiedKFold(n_splits=2, shuffle=True, random_state=123)
    scores = cross_val_score(model, meta_features, y_va_true_adjusted, cv=cv, scoring='accuracy')
    return np.mean(scores)

# コールバック関数：各試行終了時に結果を出力します。
def callback(study, trial):
    print(f"Trial {trial.number} finished with value: {trial.value}")

# Optuna スタディの作成と最適化（試行回数は n_trials=5 としています）
sampler = optuna.samplers.TPESampler(seed=123)
study = optuna.create_study(direction='maximize', sampler=sampler)
study.optimize(objective, n_trials=5, callbacks=[callback])

print("Best hyperparameters:", study.best_params)

# 最適なハイパーパラメータで最終モデルを学習・評価
final_model = lgb.LGBMClassifier(**study.best_params)
final_model.fit(meta_features, y_va_true_adjusted)
y_pred = final_model.predict(meta_features) + 1  # LightGBMは0-indexedなので元のラベルに戻す
acc = accuracy_score(y_va_true, y_pred)
print(f"Final ensemble accuracy: {acc:.4f}")
```
